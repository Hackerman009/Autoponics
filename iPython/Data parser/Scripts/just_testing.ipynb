```python
file = open("CoolTerm Capture 2021-07-01 16-59-27.txt")
data = file.readlines()
file.close()

#temperature list for graph
templ = []
for line in data:
    templ.append(float(line[6:12]))

#rint(templ[:50])
#rint(len(templ))

#humidity list for graph
huml = []
for line in data:
    huml.append(float(line[23:28]))

#rint(huml[:50])
#rint(len(huml))
```

```python
import numpy as np

def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w

#converting lists to np arrays
templ = np.array(templ)
templ = np.array(templ)

tempmean = moving_average(templ, 300)
hummean = moving_average(huml, 300)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(50, 6), dpi=80)
plt.plot(templ)
plt.ylabel('Temperature (C)')
plt.xlabel('Time (seconds)')
```

```python
def Coolterm2plot(filepath):
    
    def Coolterm2temp(filepath):
        file = open(filepath)
        data = file.readlines()
        file.close()

        #temperature list for graph
        templ = []
        for line in data:
            templ.append(float(line[26:32]))
        return templ

    def Coolterm2hum(filepath):
        file = open(filepath)
        data = file.readlines()
        file.close()

        #humidity list for graph
        huml = []
        for line in data:
            huml.append(float(line[43:49]))
        return huml

    def Coolterm2time(filepath):
        file = open(filepath)
        data = file.readlines()
        file.close()

        #humidity list for graph
        timel = []
        for line in data:
            timel.append(float(line[11:19]))
        return timel

    def moving_average(x, w):
        import numpy as np
        x = np.array(x)
        return np.convolve(x, np.ones(w), 'valid') / w
    

    def plotter(lname, lnamemean, lnamexlabel, lnameylabel):

        import matplotlib.pyplot as plt
        from matplotlib.pyplot import figure

        figure(figsize=(50, 6), dpi=240)
        plt.plot(lname)
        plt.plot(lnamemean)
        plt.xlabel(lnamexlabel)
        plt.ylabel(lnameylabel)

    #getting lists of temperature and humidity
    templ = Coolterm2temp(filepath)
    huml = Coolterm2hum(filepath)

    #getting moving averages of temperature and humidity
    tempmean = moving_average(templ, 20)
    hummean = moving_average(huml, 20)
    
    #plotting temperature and humidity
    plotter(templ, tempmean, 'Time', 'Temperature (C)')
    plotter(huml, hummean, 'Time', 'Humidity')
```
